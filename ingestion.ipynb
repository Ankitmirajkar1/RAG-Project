### PDF loader

```python
## Load pdf document

from langchain_community.document_loaders import PyPDFLoader
```

```python
file_path = "D:\\GenAI\\Agentic AI\\Test_Projects\\RAG-Project\\data\\PDF\\Rainbow-Bazaar-Return-Refund-&-Cancellation-Policy.pdf"
pdf_loader = PyPDFLoader(file_path)
```

```python
pdf_docs = pdf_loader.load()
```

```python
print(pdf_docs[0].page_content[:700])
```

### Excel loader

```python
## Load data from excel

from langchain_core.documents import Document
import pandas as pd
```

```python
df = pd.read_excel("D:\\GenAI\\Agentic AI\\Test_Projects\\RAG-Project\\data\\excel\\Mumbai_Housing Price.xlsx", sheet_name="Sheet1")
```

```python
df.head()
```

```python
excel_docs = [
    Document(page_content=row.to_json())
    for _, row in df.iterrows()
]

print(excel_docs[1].page_content)
```

### Load data from Wikipedia

```python
from langchain_community.document_loaders import WikipediaLoader
```

```python
wiki_loader = WikipediaLoader(query="List of countries by GDP", load_max_docs=1)
wiki_docs = wiki_loader.load()
```

```python
print(wiki_docs[0].page_content[:700])
```

```python
all_docs = pdf_docs + excel_docs + wiki_docs
```
